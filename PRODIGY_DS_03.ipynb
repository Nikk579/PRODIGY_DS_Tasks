Problem Statement- 
Build a decision tree classifier to predict whether a customer will purchase a product or service based on their demographic and behavioral data.
    Use a dataset such as the Bank Marketing dataset from the UCI Machine Learning Repository.

## Importing the neccessary libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv("bank-additional-full.csv", sep=';') 
df
```

```python
df_new = df.drop(["campaign","pdays","previous","poutcome","emp.var.rate","cons.price.idx","cons.conf.idx","euribor3m","nr.employed","duration","contact","default"],axis=1)
```

```python
df_new.head()
```

```python
df_new.describe
```

```python
inputs = df_new.drop(["y"],axis=1)
target= df.y
```

```python
# As the ml model better works on the numerical data we have to convert our categorical varibles into the numrical 
# we do this by using the lable encoder and one hot encoding 
from sklearn.preprocessing import LabelEncoder
le_marital = LabelEncoder()
le_job = LabelEncoder()
le_education = LabelEncoder()
le_housing = LabelEncoder()
le_loan = LabelEncoder()
le_month = LabelEncoder()
le_day_of_week = LabelEncoder()
```

```python
inputs['marital_n'] = le_marital.fit_transform(inputs['marital'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['education_n'] = le_education.fit_transform(inputs['education'])
inputs['housing_n'] = le_housing.fit_transform(inputs['housing'])
inputs['loan_n'] = le_loan.fit_transform(inputs['loan'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(["job","marital","education","housing","loan","month","day_of_week"],axis=1)
inputs_n
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs_n,target,test_size=0.2,random_state=42)
```

```python
len(x_test)
```

```python
len(x_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
clf = DecisionTreeClassifier(criterion='gini',random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf,inputs_n,target, cv=10)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print("Confusion Matrix:")
print(conf_matrix)

#classification report (precision, recall, F1-score, support)
class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)
```

## Conclusion 

The overall accuracy is 0.87 or 87%, which means that the model correctly predicted 87% of the instances in the test set.
