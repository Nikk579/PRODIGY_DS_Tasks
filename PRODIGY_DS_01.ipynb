## Visualiazing the Age, Sex Dataset with the help of histogram or bar chart

Problem Statement - Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, 
such as the distribution of ages or genders in a population.

```python
# Importing The required libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Randomly generating the data 
np.random.seed(42)

# Generating random ages between 18 and 70 for a population of 100 individuals
ages = np.random.randint(10, 70, 100)

# Generating random gender for each individual
gender_options = ['Male', 'Female']
gender = np.random.choice(gender_options, 100)

data = {'Age': ages, 'Gender': gender}

df = pd.DataFrame(data)
print(df.head())

```

```python
df.shape
```

```python
# Plotting the histogram for Age

plt.figure(figsize=(12, 4))  
plt.subplot(1, 2, 1)
plt.hist(df['Age'], bins=6,edgecolor='k', alpha=0.8, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')

# Plotting the histogram for Gender

plt.subplot(1, 2, 2)
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=['red', 'blue'], alpha=0.7)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Distribution of Gender')
plt.show()
```

## Conclusion 

We created the histogram for the Age as well as gender. We generate the data from python random module . 

Conclusion For Age- The most of the population belongs from 10 to 20 and 60 to 70 age group

Conclusion For Gender- There are approximately 60% males in the pouplation and remaining 40% are the females
